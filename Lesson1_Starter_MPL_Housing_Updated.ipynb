## 6) Linear Regression â€” np.polyfit
- a, b = np.polyfit(X, y, 1)



```python
# TODO: Regression overlay code here
x = df['SquareFeet'].values;
y = df['Price'].values
a, b = np.polyfit(x,y,1)
plt.figure(figsize=(20,10))
plt.scatter(x,y,alpha= 0.3)
xs = np.linspace(x.min(), x.max(),200)
ys = a*xs + b
plt.plot(xs, ys, linewidth= 2.5, linestyle='--')
plt.grid(True)
plt.show()
```
