# Lesson 4 Homework: Exploring Movie Insights

```python
# Import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style='whitegrid', context='talk')

# Load dataset
df = pd.read_csv("Movie_Insights.csv")

# Inspect data
print(df.info())
print(df.describe())
print(df.head())
```

### Question 1

```python
# Your code here

# Numerical columns for correlation
numerical_cols = ['Budget_Millions', 'Marketing_Spend_Millions',
                  'Revenue_Millions', 'Profit_Millions',
                  'IMDB_Rating', 'Audience_Score']

# Correlation matrix
corr = df[numerical_cols].corr()

# Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Heatmap  Exploring Relationships", fontsize=16)
plt.show()
```

### Question 2

```python
# Your code here

# Theme
sns.set_theme(style='whitegrid', context='talk')

# Create bubble plot
plt.figure(figsize=(12, 8))
scatter = sns.scatterplot(
    data=df,
    x='Budget_Millions',
    y='Revenue_Millions',
    hue='IMDB_Rating',          # Color by IMDB rating
    palette='viridis',          # Color palette
    size='Audience_Score',      # Bubble size by audience score
    sizes=(50, 500),            # Min and max bubble sizes
    alpha=0.7,                  # Slight transparency
    edgecolor='black'
)

# Add title and labels
plt.title("Budget vs Revenue with Ratings and Audience Scores", fontsize=16)
plt.xlabel("Budget (Millions)")
plt.ylabel("Revenue (Millions)")

# Move legend outside
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)

plt.show()
```

### Question 3

```python
# Your code here

sns.set_theme(style='whitegrid', context='talk')

plt.figure(figsize=(12, 6))
sns.barplot(
    data=df,
    x='Studio',
    y='Profit_Millions',
    hue='Studio',           # Fix for palette warning
    legend=False,           # Hide duplicate legend
    errorbar=None,          # Replaces deprecated ci=None
    palette='Set3'
)

plt.title("Average Profit by Studio", fontsize=16)
plt.xlabel("Studio")
plt.ylabel("Average Profit (Millions)")
plt.xticks(rotation=45)
plt.show()
```
