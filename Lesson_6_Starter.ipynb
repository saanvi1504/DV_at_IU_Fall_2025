```python
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import numpy as np

df = pd.read_csv('Global_Space_Missions_Final_1200.csv')
df['Launch_Year'] = pd.to_datetime(df['Launch_Date']).dt.year
df['Launch_Month'] = pd.to_datetime(df['Launch_Date']).dt.month
df['Outcome'] = np.where(df['Success'].astype(str).str.lower().isin(['true','1','yes','y']), 'Success', 'Failure')

df.head()
```

## Visualization 2

```python
# Map common launch sites to coordinates (approx) — minimal runtime mapping
site_latlon = {
    'Cape Canaveral': (28.3922, -80.6077),
    'Kennedy Space Center': (28.5729, -80.6490),
    'Vandenberg': (34.7420, -120.5724),
    'Baikonur': (45.9647, 63.3050),
    'Kourou': (5.2360, -52.7680),
    'Tanegashima': (30.4010, 130.9740),
    'Satish Dhawan': (13.7199, 80.2304),
    'Jiuquan': (40.9606, 100.2983),
    'Wenchang': (19.6145, 110.9510),
    'Xichang': (28.2460, 102.0260),
    'Plesetsk': (62.9256, 40.5770)
}
df['lat'] = df['Launch_Site'].map(lambda s: site_latlon.get(str(s), (np.nan, np.nan))[0])
df['lon'] = df['Launch_Site'].map(lambda s: site_latlon.get(str(s), (np.nan, np.nan))[1])
geo = df.dropna(subset=['lat','lon'])

fig3 = px.scatter_geo(
    geo,
    lat='lat', lon='lon', color='Agency', size='Mission_Cost_Million',
    hover_name='Mission_Name', hover_data=['Launch_Site','Country','Launch_Year'],
    animation_frame='Launch_Year',
    projection='natural earth',
    title='Animated Globe: Launch Sites by Year (Bubble = Mission Cost)'
)
fig3.update_layout(legend_title_text='Agency')
fig3.show()
```

## Visualization 7

```python
# Build Sankey from grouped flows
flows = df.groupby(['Country','Mission_Type','Outcome']).size().reset_index(name='Count')

# Unique nodes across all stages
countries = flows['Country'].unique().tolist()
types = flows['Mission_Type'].unique().tolist()
outcomes = flows['Outcome'].unique().tolist()

labels = countries + types + outcomes

# Index helpers
index = {name:i for i,name in enumerate(labels)}

# Country -> Mission Type links
ct = df.groupby(['Country','Mission_Type']).size().reset_index(name='Count')
sources1 = ct['Country'].map(index).tolist()
targets1 = ct['Mission_Type'].map(lambda x: index[x]).tolist()
values1 = ct['Count'].tolist()

# Mission Type -> Outcome links
to = df.groupby(['Mission_Type','Outcome']).size().reset_index(name='Count')
sources2 = to['Mission_Type'].map(lambda x: index[x]).tolist()
targets2 = to['Outcome'].map(lambda x: index[x]).tolist()
values2 = to['Count'].tolist()

fig10 = go.Figure(data=[go.Sankey(
    node=dict(pad=20, thickness=20, line=dict(color='black', width=0.5), label=labels),
    link=dict(
        source=sources1+sources2,
        target=targets1+targets2,
        value=values1+values2
    )
)])
fig10.update_layout(title_text='Sankey: Country → Mission Type → Outcome', font_size=12)
fig10.show()
```
